<h2>Time and accuracy analysis of polynomial and exponetial time algorithms.
</h2>

```python
from __future__ import print_function, division
import string
import numpy as np
import time
import random
from sklearn.linear_model import LinearRegression
from matplotlib import pylab
import seaborn as sns

from LinearSpace import LinearSpaceAlignment
from EditDistanceDP_OOP import DynamicProgrammingAlignment
from KstripModel import K_Strip_Model
from GreedyMy import Greedy
from Recursive import Recursive
from BranchBound import BranchBound
```

<h4>In the cell below we measure the running time of polynomial-time algorithms. We use python buit-in time.clock() method which gives CPU time at the moment of call. Running time is measured as time of the finish of the algorithm minus time of the start of the algorithm. For each string lenght, from 1 to 50, 20 iterations is made to measure the average performance. We evaluate algorithms on random strings from ASCII set.</h4>

```python
running_times = {"DynamicProgramming":[],
                    "Recursive":[],
                    "BranchAndBound":[],
                    "Greedy":[],
                    "DivideAndConquer":[],
                    "kstripModel":[]}
def test_time(): 
    for i in range(1,50): 
        _dynamic = []
        _kstripBest = []
        _kstripModel = []
        _divide = []
        _greedy = []
        for _ in range(20):
            test_string1 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
            test_string2 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
            
            #dynamic 
            
            start = time.clock()
            algo = DynamicProgrammingAlignment(test_string1, test_string2)
            algo.compute_matrix()
            algo.print_alignment()
            stop = time.clock()
            _dynamic.append(stop-start)
            
            #divide
            start = time.clock()
            algo = LinearSpaceAlignment(test_string1,test_string2)
            algo.align()
            algo.print_alignment()            
            stop = time.clock()
            _divide.append(stop-start)
            
            #kstrip linear model 
            start = time.clock()
            algo = K_Strip_Model(test_string1, test_string2)
            algo.k_strip()
            algo.print_alignment()
            stop = time.clock()
            _kstripModel.append(stop-start)
            #greedy
            start = time.clock()
            algo = Greedy(test_string1,test_string2)
            algo.ed_greedy()
            algo.align()
            stop = time.clock()
            _greedy.append(stop-start)
        running_times["DynamicProgramming"].append(np.mean(_dynamic))
        running_times["DivideAndConquer"].append(np.mean(_divide))
        running_times["kstripModel"].append(np.mean(_kstripModel))
        running_times["Greedy"].append(np.mean(_greedy))
test_time()
```

<h3>We repeat the same process for the exponential time algorithms. In this case due to reasonoble performance time restrictions, we bound the string length to 9 and 15, for Recursive and Branch And Bound respectively. Values 9 and 15 are found empirically</h3>

```python
running_times["Recursive"] = []
running_times["BranchAndBound"] = []
def test_time_recursive():
    for i in range(1,16):
        _recursive = []
        _branch = []
        for j in range(20): 
            test_string1 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
            test_string2 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))

            #branch and bound
            start = time.clock()
            algo = BranchBound(test_string1,test_string2)
            algo.align(test_string1,test_string2)
            stop =  time.clock()
            _branch.append(stop-start)
            #pure recursive
            if i<=9:
                start = time.clock()
                algo = Recursive()
                algo.align(test_string1,test_string2)
                stop = time.clock()
                _recursive.append(stop-start)

                    
        running_times["BranchAndBound"].append(np.mean(_branch))
        if i <= 9: 
            running_times["Recursive"].append(np.mean(_recursive))
        if i > 9 :
            running_times["Recursive"].append(None)
            
test_time_recursive()
            
```

```python
### We plot the average performance for polynomial time algorithms 

string_lenghts = [i for i in range(1,50)]

fig = pylab.gcf()
fig.set_size_inches(18.5, 10.5)
pylab.plot(string_lenghts, running_times["DynamicProgramming"], 'r-', label= "Dynamic Programming")
pylab.plot(string_lenghts, running_times["DivideAndConquer"], 'b-', label= "Divide And Conquer")
pylab.plot(string_lenghts, running_times["kstripModel"], 'c-', label= "K-Strip")
pylab.plot(string_lenghts, running_times["Greedy"], 'k-', label= "Greedy")
pylab.legend(loc='upper left',prop={'size': 15})
pylab.title("Average running time for the random strings of length 1 to 50 for polynomial algorithms")
pylab.xlabel("String Length")
pylab.ylabel("Runing time in seconds")
pylab.savefig("running_time_poly.png")
pylab.show()
```

```python
### We plot the average performance for exponential time algorithms 


string_lenghts = [i for i in range(1,16)]

fig = pylab.gcf()
fig.set_size_inches(18.5, 10.5)
pylab.plot(string_lenghts, running_times["Recursive"], 'r-', label= "Recursive")
pylab.plot(string_lenghts, running_times["BranchAndBound"], 'b-', label= "Branch And Bound")
pylab.legend(loc='upper left',prop={'size': 15})
pylab.title("Average running time for the random strings of lenght 1 to 15 for exponential algorithms. (Recursive is stopped after string lenght 9.)")
pylab.xlabel("String Lenght")
pylab.ylabel("Runing time in seconds")
pylab.savefig("running_time_expo.png")
pylab.show()
```

```python
max_len = 100
greedy_accuracy = []
k_strip_accuracy = []
greedy_string_len = [i for i in range(1,max_len)]
def measure_accuracy():
    for i in range(1,max_len):
        _greedy_ed = []
        _dynamic_ed = []
        _kstrip_ed = []
        for _ in range(20):
            test_string1 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
            test_string2 =  ''.join(random.choice(string.ascii_uppercase) for _ in range(i))
            
            #dynamic 
            algo = DynamicProgrammingAlignment(test_string1, test_string2)
            algo.compute_matrix()
            edit_distance_dp = algo.get_edit_distance()
            _dynamic_ed.append(edit_distance_dp)

            #greedy
            algo = Greedy(test_string1,test_string2)
            edit_distance_greedy, _ = algo.ed_greedy()
            _greedy_ed.append(edit_distance_greedy)
            #kstrip model
            algo = K_Strip_Model(test_string1, test_string2)
            edit_distance_k_strip = algo.k_strip()[0]
            _kstrip_ed.append(edit_distance_k_strip)            
        greedy_accuracy.append(np.mean(_dynamic_ed)/np.mean(_greedy_ed))
        k_strip_accuracy.append(np.mean(_dynamic_ed)/np.mean(_kstrip_ed))
            
measure_accuracy()
```

```python
# we use liniear regression to generalize the trend of accuracy graph
x = np.ones(shape=(99,2))
x[:,1] = np.arange(99)
lr = LinearRegression()
lr.fit(x,greedy_accuracy)
line_greedy = lr.predict(x)
lr.fit(x,k_strip_accuracy)
line_k_strip = lr.predict(x)
```

```python
# we plot the accuracy (0 <= Accuracy <= 1) of approximate algorithms as EditDistance/Approx.EditDistance, where 1 is exact edit distance

fig = pylab.gcf()
fig.set_size_inches(18.5, 10.5)
pylab.plot(greedy_string_len, greedy_accuracy, 'r-', label= "Greedy")
pylab.plot(greedy_string_len, line_greedy, 'r--', label= "Greedy Trend")
pylab.plot(greedy_string_len, k_strip_accuracy, 'c-', label= "K Strip")
pylab.plot(greedy_string_len, line_k_strip, 'c--', label= "K Strip Trend")
pylab.legend(loc='lower left',prop={'size': 15})
pylab.title("Average accuracy of approximate algorithms for the random strings of length 1 to 100.")
pylab.xlabel("String Length")
pylab.ylabel("Accuracy (Dynamic Programming Accuracy/Approximate algorithm Accuracy)")
pylab.savefig("accuracy_aproximate.png")

pylab.show()
```
